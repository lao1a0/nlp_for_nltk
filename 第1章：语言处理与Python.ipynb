计算词频，以百分比表示


```python
def percent(word, text):
     return 100*text.count(word)/len([w for w in text if w.isalpha()])
percent(",", text1)
```


计算文本词汇量


```python
def vocab_size(text):
    return len(set(w.lower() for w in text if w.isalpha()))
vocab_size(text1)
```
