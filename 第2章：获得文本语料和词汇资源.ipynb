1. `brown.words(fileids=['cg22'])`：这个函数通过指定文件标识符（fileid）来获取数据。在布朗语料库中，每个文本都被分配了一个唯一的标识符。这里的`cg22`是其中一个文件的标识符，它代表的是纯文学文体（belles_lettres）的一个具体文档。使用`fileids`参数时，你可以直接访问这个特定文件中的所有单词。

2. `brown.words(categories='news')`：这个函数通过指定类别（category）来获取数据。布朗语料库将文本分为不同的类别，如新闻（news）、社论（editorial）、小说（fiction）等。当你使用`categories`参数并指定`'news'`时，NLTK会返回所有被标记为新闻类别的文档中的单词。

总结来说，`fileids`参数允许你访问特定文件的单词，而`categories`参数允许你访问特定类别下所有文件的单词。两者都是从布朗语料库中提取数据的有效方式，但它们关注的粒度不同。


```python
brown.sents(categories=['news', 'editorial', 'reviews'])
```

```python
import nltk
from nltk.corpus import brown
news_text = brown.words(categories='news')
fdist = nltk.FreqDist(w.lower() for w in news_text)
fdist
```

```python
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
        print(m + ':', fdist[m])
```

```python
cfd = nltk.ConditionalFreqDist(
    (genre, word)
    for genre in brown.categories()
    for word in brown.words(categories=genre))
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditions=genres, samples=modals)
```

# 2 条件频率分布

## 2.1 条件和事件

每个**配对pairs**的形式是：(条件, 事件)。如果我们按文体处理整个布朗语料库，将有15 个条件（每个文体一个条件）和1,161,192 个事件（每一个词一个事件）。
- 频率分布计算观察到的事件，如文本中出现的词汇。
- 条件频率分布需要给每个事件关联一个条件，所以不是处理一个词序列①，而是要处理一系列配对序列②

```python
text = ['The', 'Fulton', 'County', 'Grand', 'Jury', 'said', ]  # ①
pairs = [('news', 'The'), ('news', 'Fulton'), ('news', 'County'), ] # ②
```

利用bigrams制作生成模型：


```python
sent = ['In', 'the', 'beginning', 'God', 'created', 'the', 'heaven','and', 'the', 'earth', '.']
nltk.bigrams(sent)
```

例 2-1. 产生随机文本：此程序获得了《创世纪》文本中所有的双连词，然后构造一个条件频率分布来记录哪些词汇最有可能会跟在给定词的后面，例如：living 后面最可能的词是 creature。generate_model()函数使用这些数据和种子词来产生随机文本

```python

def generate_model(cfdist, word, num = 15):
     for i in range(num):
             print(word,end = " ")
             word = cfdist[word].max()

text = nltk.corpus.genesis.words("english-kjv.txt")
bigrams = nltk.bigrams(text)
cfd = nltk.ConditionalFreqDist(bigrams)
cfd
```

```python
cfd['living']
```

```python
cfd['creature']
```

```python
generate_model(cfd, 'living')
```

在例 2-1 中，将每个词作为一个条件，对于每个词都有效地依据后续词的创建频率分布。函数 generate_model()包含简单的循环以生成文本。在调用函数时，选择一个词（如 living”）作为初始内容。进入循环后，输入变量 word 的当前值，重新设置 word 为上下文中最可能的标识符（使用 max()）。下一次进入循环时，将这个词作为新的初始内容。通过检查输出可以发现，这种简单的文本生成方法往往会在循环中卡住。另一种方法是从可用的词汇中随机选择下一个词。

## 4.1 词汇列表语料库
**词汇语料库**是Unix 中的/usr/share/dict/words文件，被一些拼写检查程序使用。我们可以用它来寻找文本语料中不寻常的或拼写错误的词汇。


```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab - english_vocab
    return sorted(unusual)

unusual_words(nltk.corpus.gutenberg.words('austen-sense.txt'))
```

```python
unusual_words(nltk.corpus.nps_chat.words())
```

**停用词语料库**，就是那些高频词汇，如the，to和also，我们有时在进一步的处理之前想要将它们从文档中过滤。

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / len(text)
content_fraction(nltk.corpus.reuters.words()) # 来计算文本中不包含在停用词列表中的词所占的比例
```

```python
puzzle_letters = nltk.FreqDist('egivrvonl') # 注意，这里如果是字符串'egivrvonl'的话，给出的就是每个字母的频数
puzzle_letters
```

```python
obligatory = 'r'
wordlist = nltk.corpus.words.words()
wordlist
```

```python
[w for w in wordlist if len(w) >= 6 
                      and obligatory in w 
                      and nltk.FreqDist(w) <= puzzle_letters]
```

用于从一个单词列表（`wordlist`）中筛选出满足特定条件的单词。：

1. `if len(w) >= 6`：只有长度至少为6个字符的单词才会被考虑。

2. `and obligatory in w`：这个条件检查`obligatory`字符串是否是单词`w`的子串。只有当`obligatory`是`w`的一部分时，`w`才会被保留。

3. `and nltk.FreqDist(w) <= puzzle_letters`： 这里涉及到NLTK库中的`FreqDist`函数，它用于计算单词`w`中各个字母的频率分布。`puzzle_letters` 定义了每个字母在谜题中允许出现的最大次数。这个条件检查单词`w`中每个字母的频率是否不超过`puzzle_letters`中定义的限制。

综合来看，这段代码的作用是：从`wordlist`中筛选出至少包含6个字符、包含特定子串`obligatory`，并且每个字符的出现频率不超过`puzzle_letters`中定义的限制的单词。用于解决单词拼图或者填字游戏，其中`obligatory`是必须包含的单词片段，而`puzzle_letters`定义了每个字母的使用限制。

这段文本提供了 WordNet 中 "mint" 这个词的六个不同同义词集（Synset）的定义，每个同义词集代表 "mint" 的一个不同意义。下面是每个同义词集的编号和定义：

1. `batch.n.02`： 这个定义与 "mint" 这个词无关，可能是一个错误或者上下文混淆。`batch` 指的是一大批或大量的数量或程度，通常后面会跟 "of"。

2. `mint.n.02`：这个同义词集指的是薄荷属（Mentha）的任何一种北温带植物，这种植物有芳香的叶子和小型的淡紫色花朵。

3. `mint.n.03`：这个同义词集指的是薄荷科（Lamiaceae）的任何一种植物。

4. `mint.n.04`：这个同义词集指的是新鲜的或糖渍的薄荷叶。

## 5.3 语义相似度


```python
right = wn.synset('right_whale.n.01')
orca = wn.synset('orca.n.01') # 虎鲸
minke = wn.synset('minke_whale.n.01') # 小鱼
tortoise = wn.synset('tortoise.n.01') # 龟
novel = wn.synset('novel.n.01') # 小说
```

```python
right.lowest_common_hypernyms(minke) # 须鲸
```


`right.lowest_common_hypernyms(minke)` 这个表达式是在WordNet中查找两个同义词集（Synset）之间的最低共同上位词（hypernym）。

在这个具体的例子中，`right` 和 `minke` 分别代表两个不同的同义词集，`right` 对应的是 "right whale"（露脊鲸），而 `minke` 对应的是 "minke whale"（小须鲸）。这个函数返回的是这两个词集共有的最具体的上位词，也就是在WordNet的层次结构中，这两个词集最近的共同祖先。

根据搜索结果，`right.lowest_common_hypernyms(minke)` 的结果是 `[Synset('baleen_whale.n.01')]`，这意味着 "right whale" 和 "minke whale" 的最低共同上位词是 "baleen whale"（须鲸）。这表明在WordNet的层次结构中，"right whale" 和 "minke whale" 都属于 "baleen whale" 这个更广泛的类别。

```python
right.lowest_common_hypernyms(orca)
```

```python
right.lowest_common_hypernyms(tortoise) # 脊椎动物
```

```python
right.lowest_common_hypernyms(novel) # 实体
```

WordNet 同义词集定义了类似的方法来进行深入的观察。例如：

path_similarity 基于上位词层次结构概念中相互关联的最短路径下，在 0～1 范围内的相似度（两者之间没有路径就返回−1）。同义词集与自身比较将返回 1。

考虑以下的相似度：露脊鲸与小须鲸、逆戟鲸、乌龟以及小说。从海洋生物的语义空间转移到非生物时，数字越接近于0
